# GRAYSCALE CONVERSION

TRAINING SET

```python
import cv2

import os,glob

from os import listdir,makedirs

from os.path import isfile,join
path = r'C:\Users\raajshree\Downloads\500'
dstpath = r'C:\Users\raajshree\Downloads\500'
try:
    makedirs(dstpath)
except:
    print ("Directory already exist, images will be written in same folder")
files = list(filter(lambda f: isfile(join(path,f)), listdir(path)))
for image in files:
    try:
        img = cv2.imread(os.path.join(path,image))
        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        dstPath = join(dstpath,image)
        cv2.imwrite(dstPath,gray)
    except:
        print ("{} is not converted".format(image))
for fil in glob.glob("*.jpg"):
    try:
        image = cv2.imread(fil)
        gray_image = cv2.cvtColor(os.path.join(path,image), cv2.COLOR_BGR2GRAY) # convert to greyscale
        cv2.imwrite(os.path.join(dstpath,fil),gray_image)
    except:
        print('{} is not converted')

```

TEST CODE

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.144])

img = mpimg.imread('test-image-cover.jpeg')

gray = rgb2gray(img)

plt.imshow(gray, cmap = plt.get_cmap('gray'))

plt.savefig('test-image-cover1.jpeg')
plt.show()
```

# EDGE DETECTION

TRAINING SET

```python
import cv2
import numpy as np
import glob
def auto_canny(image, sigma=0.33):
    v = np.median(image)
    lower = int(max(0, (1.0 - sigma) * v))
    upper = int(min(255, (1.0 + sigma) * v))
    return cv2.Canny(image, lower, upper)
images = [cv2.imread(file,0) for file in
glob.glob(r"C:\Users\raajshree\Downloads\500/*.jpg")]
number = 0
for image in images:
    canny = auto_canny(image)
    cv2.imwrite(r'C:\Users\raajshree\Downloads\500/canny_{}.png'.format(number), canny)
    number += 1
```

TEST CODE

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('20.jpg',0)
edges = cv2.Canny(img,100,200)

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.savefig('test-img2.jpeg')
plt.show()
```
